```python
def pattern_matching(pattern, genome):
    '''
    Input: Two strings, Pattern and Genome.
    Output: A collection of space-separated integers specifying all starting positions where Pattern appears
            as a substring of Genome
    '''
    starting_positions = ""
    position = 0
    
    while True:
        position = genome.find(pattern, position)
        
        if not position == -1:
            starting_positions = starting_positions + " " + str(position)
            position += 1
        else:
            break

    return starting_positions.strip()
```

```python
pattern = 'ATAT'
genome = 'GATATATGCATATACTT'
```

```python
print pattern_matching(pattern, genome)
```
